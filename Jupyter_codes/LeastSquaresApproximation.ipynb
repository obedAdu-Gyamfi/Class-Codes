```python
import numpy as np
from matplotlib import pyplot as plt

if __name__ == "__main__":
    class LeastSquare:
        def __init__(self, x, y):
            self.x = x
            self.y = y
            self.n = len(x)

        def linear(self):
            SumX = np.sum(self.x)
        SumY = np.sum(self.y)
        SumXX = 0
        for i in self.x:
            SumXX += i**2
        SumXY = 0
        for (i,j) in zip(self.x,self.y):
            SumXY += i * j
        Slope = (SumXY - (SumX * SumY)/self.n )/(SumXX - (SumX**2)/self.n)
        Intercept = (SumY)/ self.n - (Slope * (SumX)/self.n)
        if Intercept >= 0:
            print("Y = {:.4f}X + {:.4f}".format(Slope, Intercept))
        else:
            print("Y = {:.4f}X {:.4f}".format(Slope, Intercept))
        plt.scatter(self.x,self.y)
        plt.grid('on')
        x_fit = np.linspace(min(self.x), max(self.x),100)
        y_fit = Slope*x_fit + Intercept
        plt.plot(x_fit,y_fit, color='red')
        plt.title('Least Square Linear Approximation')
        plt.xlabel('x')
        plt.ylabel('y')
        plt.show()
        
    def polynomial(self):
        sum_x = np.sum(self.x)
        sum_y = np.sum(self.y)
        sum2 = np.sum(np.square(self.x))
        sum3 = np.sum(np.power(self.x,3))
        sum4 = np.sum(np.power(self.x,4))
        sumxy = 0
        for (i,j) in zip(self.x, self.y):
            sumxy += (i * j)
        sum2y = np.sum(np.square(self.x)* self.y)
        A = np.array(([self.n,sum_x,sum2],[sum_x,sum2,sum3],[sum2,sum3,sum4]))
        print(A)
        B_ = np.array(([sum_y,sumxy,sum2y]))
        B = B_.transpose()
        print(B)
    def exponential(self):
        lny = np.log(self.y[:])
        sum_lny = np.sum(lny)
        sum_x = np.sum(self.x)
        sum_Xlny = 0
        sum_xX = 0
        for i in self.x:
            sum_xX += i**2
        for (i,j) in zip(self.x,lny):
            sum_Xlny += (i * j)

        slope =((self.n*sum_Xlny) - (sum_x * sum_lny))/(self.n*sum_xX - (sum_x**2)) 
        Intercept = np.exp((sum_lny - (slope * sum_x))/self.n)

        x_fit = np.linspace(min(self.x), max(self.x), 1000)
        y_fit = Intercept*np.exp(slope*x_fit)
        print("Y = {:4f}exp({:4f}x)".format(Intercept, slope))

        plt.scatter(self.x, self.y, c='red', marker='o', s=50)
        plt.grid('on')
        plt.plot(x_fit, y_fit, color='blue')
        plt.title('Least Square Exponential Approximation')
        plt.xlabel('x')
        plt.ylabel('y')
        plt.show()
        
X1 = [0,0.25,0.5,0.75,1]
Y1 = [1,1.2840,1.6487,2.1170,2.7183]
L1 = LeastSquare(X1,Y1)
L1.exponential()
L1.linear()
#L1.polynomial()
```
